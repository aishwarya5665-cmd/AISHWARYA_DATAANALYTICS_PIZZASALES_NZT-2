**c. What data cleaning steps would you apply?**

```python
#Removing duplicates
df.drop_duplicates(inplace=True)
```

```python
#Time-based Columns
df["Week"] = df["order_date"].dt.isocalendar().week
df["Month"] = df["order_date"].dt.month
df["Day"] = df["order_date"].dt.day_name()
```

**STEP 2-DESCRIPTIVE STATISTICS**

```python
#Parse dates
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df['Week'] = df['order_date'].dt.isocalendar().week
```

**a. Find the total revenue for each product.**

```python
#Total Revenue for each product
revenue_per_product = df.groupby("pizza_name")["total_price"].sum()
```

**b. Compute average weekly units sold for Product A and Product B.**

```python
#Average Weekly Units Sold for Product A and Product B
productA = "The Classic Deluxe Pizza"
productB = "The Five Cheese Pizza"

weekly_units = df.groupby(["Week", "pizza_name"])["quantity"].sum().reset_index()

avgA = weekly_units[weekly_units["pizza_name"]==productA]["quantity"].mean()
avgB = weekly_units[weekly_units["pizza_name"]==productB]["quantity"].mean()
```

**c. Identify which week had the highest sales overall.**

```python
#Extract Week number
df["Week"] = df["order_date"].dt.isocalendar().week
```

```python
# Calculate total sales per week
weekly_sales = df.groupby("Week")["total_price"].sum()
```

```python
#Week with Highest Sales Overall
weekly_sales = df.groupby("Week")["total_price"].sum()
highest_week = weekly_sales.idxmax()
highest_week_sales = weekly_sales.max()

revenue_per_product, avgA, avgB, highest_week, highest_week_sales
```

**STEP 3-DATA VISUALIZATION**

**a. Line chart of weekly units sold for each product.**

```python
import matplotlib.pyplot as plt
```

```python
weekly_units = df.groupby(["Week", "pizza_name"])["quantity"].sum().reset_index()

plt.figure(figsize=(10,6))
for product in weekly_units["pizza_name"].unique():
    product_data = weekly_units[weekly_units["pizza_name"] == product]
    plt.plot(product_data["Week"], product_data["quantity"], label=product)

plt.xlabel("Week")
plt.ylabel("Units Sold")
plt.title("Weekly Units Sold per Product")
plt.legend()
plt.show()
```

**b. Bar chart showing total revenue per product.**

```python
revenue_per_product = df.groupby("pizza_name")["total_price"].sum()

plt.figure(figsize=(10,6))
plt.bar(revenue_per_product.index, revenue_per_product.values)
plt.xticks(rotation=90)
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.title("Total Revenue per Product")
plt.show()
```

**c. Scatter plot: Marketing Spend vs Units Sold.**

```python
#Create synthetic Marketing Spend column
np.random.seed(42)
df["Marketing_Spend"] = df["quantity"] * np.random.uniform(5,15)

plt.figure(figsize=(8,6))
plt.scatter(df["Marketing_Spend"], df["quantity"])
plt.xlabel("Marketing Spend")
plt.ylabel("Units Sold")
plt.title("Marketing Spend vs Units Sold")
plt.show()
```

**STEP 4- DATA ANALYSIS**

**a. Does higher marketing spend correlate with more units sold?**

```python
#create synthetic marketing spend
np.random.seed(42)
df["Marketing_Spend"] = df["quantity"] * np.random.uniform(5,15)

# compute correlation
corr = df["Marketing_Spend"].corr(df["quantity"])
corr
```

**b. Which product shows greater growth over the four weeks?**

```python
## Weekly units per product
weekly_units = df.groupby(['Week','pizza_name'])['quantity'].sum().reset_index()
```

```python
##Filter first 4 weeks
first4 = weekly_units[weekly_units['Week'].isin([1,2,3,4])]

growth = first4.pivot(index='pizza_name', columns='Week', values='quantity').fillna(0)
growth['Growth'] = growth.get(4,0) - growth.get(1,0)

growth_sorted = growth['Growth'].sort_values(ascending=False)
growth_sorted.head()
```
